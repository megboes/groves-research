```python
#importing networkx
import networkx as nx
#importing matplotlib.pyplot
import matplotlib.pyplot as plt

def generate_random_planar_graph(n):
    """Generates a random planar graph with n nodes."""

    while True:
        G=nx.erdos_renyi_graph(n, 0.5) # Generate a random graph, each edge occurs w/ prob .5
        if nx.check_planarity(G)[0]:  # Check if it's planar
            return G

G = generate_random_planar_graph(9) # change number to desired number of vertices.
nx.draw(G, with_labels =True)
```
