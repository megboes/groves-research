```python
import numpy as np
import matplotlib.pyplot as plt
import math
import sympy as sp

T1 = np.array([[math.sqrt(3)+3, math.sqrt(3)-3, -2*math.sqrt(3)],
              [math.sqrt(3)-3, math.sqrt(3)+3, -2*math.sqrt(3)]])
T = (1/6)*T1
print(T)
V=np.array([[2],[3],[4]])
print(V)
D = np.matmul(T, V)
print(D)

fig, ax = plt.subplots()
ax.plot([1, 3], [2, 4],color='green', marker='o')
ax.text(1, 2, f'({1},{2})', verticalalignment='bottom', horizontalalignment='right', fontsize=9)
plt.show()

print("This is D[0][0]")
print(D[0][0])
```
